```python
import requests
import json
```

```python
ksqldb_url = "http://localhost:8088/query"
headers = {'Content-Type':'application/vnd.ksql.v1+json; charset=utf-8'}
query={'ksql':'SELECT ROWKEY, CREATEDAT, ID, TEXT, GEOLOCATION, PLACE, USER FROM TWEETS EMIT CHANGES  LIMIT 100;'}
count = 0
while True:
    r = requests.post(ksqldb_url, data=json.dumps(query), headers=headers)
    with open(f'tweets{str(count)}.json', 'w') as outfile:
        json_object = r.json()
        outfile.write(json_object)
    count += 1

    
```
